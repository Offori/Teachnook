```python
#Linear regression-supervised learning technique(data is weel labled)
#dataset-area vs prices
#dataset-https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv
#dataset-data raw datafarme-tabled data
```

```python
#1.take the data and create dataframe
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv')
df
# imaginary story-imagine a real estate property comes to you and gives you the above dataset and says to create a 
# ML model which could predict the property rices using the dataset which I gave
```

```python
#3.data visualiation-creatin' graph
import matplotlib.pyplot as plt
#plt.plot(x_axis,y_axis)
plt.plot(df['Area'],df['Prices'])
```

```python
#4.divide the i/p into and o/p
# so i/p-area-always 2 dimensional array
# o/p-prices-always 1 dimensional array
x=df.iloc[0:6,0:1].values#iloc[rows,cols]/.values convert my data into array
x
```

```python
# o/p is always 1 dimensionl
# if ever we want out array to be 1d then do not write: in the colum's place
y=df.iloc[:,1].values
y
```

```python
#7.run a classifier/regresser/clasterer(apply algorithm)
from sklearn.linear_model import LinearRegression#package-colection of library, Linearregression-Library
model=LinearRegression()
```

```python
#fit the model
model.fit(x,y)#linearregression.fit(x,y) we are mapping all the i/p and o/p(plotting all the values of i/p and o/p)
# in the imaginary graph of linearregression we're plotting all the value of i/p(x) and o/p(y) of linear regression
```

```python
# predict the o/p
y_pred=model.predict(x)#using the i/p valuse we predict the o/p
# y_pred predict o/p
y_pred
```

```python
# so now we compare our predict o/p we the actual o/p(y)
y#actual o/p values
```

```python
# now when we compare y_pred and y values we come to know that their is a huge difference in ou predict values
#this does not that our model is inaccurate, it only means that our model is nonlinear
```

```python
# just to check if our if our model has predicted properly or not, we have a cross verification technique
# y=mx+c
# m-slope
# c-constant or y-intercept
# x-independent variable
# y-dependent variable
```

```python
m=model.coef_#m-slope
m
```

```python
c=model.intercept_#y-intercept(c)
c
```

```python
#y=mx+c
m*1400+c#we're just doing individual prediction for 1400 sqft
```

```python
#individual prediction
model.predict([[1400]])
```

```python
# visualization-best fit line
plt.scatter(x,y)
plt.plot(x,y_pred,c='orange')
```
