```python
fname_pclass=titanicDF['Pclass'].unique()
fname_sex=titanicDF['Sex'].unique()
fname_pclass,fname_sex
```

```python
fsize_pclass=titanicDF.groupby(['Pclass'],sort=False).size()
fsize_sex=titanicDF.groupby(['Sex'],sort=False).size()
fsize_pclass,fsize_sex
```

```python
import matplotlib.pyplot as plt
plt.bar(fname_pclass,fsize_pclass,color=['red','yellow','green'])#plt.bar(fname_sex,fsize_sex)
```
